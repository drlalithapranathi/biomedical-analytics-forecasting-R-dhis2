```r
library(XML)
library(ggplot2)

# Read the XML file
xmlfile <- xmlParse("563-ws-training (1).xml")

# Extract <event> nodes
events <- getNodeSet(xmlfile, "//event")

# Extract vectors from attributes
timestamps <- sapply(events, function(x) xmlAttrs(x)[["ts"]])        # time attribute
values     <- as.numeric(sapply(events, function(x) xmlAttrs(x)[["value"]]))  # glucose values

# Make dataframe
data <- data.frame(
  timestamp = as.POSIXct(timestamps, format="%d-%m-%Y %H:%M:%S"),
  glucose   = values
)
data <- data[order(data$timestamp), ]   # sort by time
```

```r
plot(data$timestamp, data$glucose, type="l", col="blue",
     main="Glucose Over Time (Base R Line Chart)",
     xlab="Time", ylab="Glucose (mg/dL)")
```

```r
ggplot(data, aes(x=timestamp, y=glucose)) +
  geom_line(color="orange") +
  labs(title="Glucose Over Time (ggplot2 Line Chart)",
       x="Time", y="Glucose (mg/dL)") +
  theme_minimal()
```

```r
ggplot(data, aes(x = glucose)) +
  geom_histogram(aes(y = after_stat(density)),   # scale histogram to density
                 binwidth = 10, fill = "blue", color = "black") +
  geom_density(linewidth = 1, color = "red") +   # add density curve
  labs(title = "Distribution of Glucose Values (Histogram + Density)",
       x = "Glucose (mg/dL)", y = "Density") +
  theme_minimal()
```
